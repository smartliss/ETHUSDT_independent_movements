```python
import numpy as np
import matplotlib.pyplot as plt

walk = np.cumsum(np.random.randn(100000))

# 
plt.figure(figsize=(10, 6))
plt.plot(walk)
plt.grid(True)
plt.show()

diff_ts = np.diff(walk)

plt.figure(figsize=(10, 6))
plt.plot(diff_ts)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Generate two cointegrated time series
np.random.seed(42)
n = 1000
x = np.cumsum(np.random.randn(n))
y = 7*x + 10 + np.random.randn(n)*3 

# Plotting
plt.figure(figsize=(14, 6))

# Plot time series to visualize cointegration
plt.plot(x, label='X')
plt.plot(y, label='Y')
plt.title('Two Cointegrated Time Series')
plt.legend()
plt.show()

# Augmented Dickey-Fuller test to check non-stationarity of individual series
print("ADF Test for X:", adfuller(x)[1])  # p-value > 0.05 indicates non-stationary
print("ADF Test for Y:", adfuller(y)[1])  # p-value > 0.05 indicates non-stationary

diff = y - 7*x
plt.figure(figsize=(14, 6))
plt.plot(diff)
plt.grid(True)
plt.show()

print("ADF Test for Diff:", adfuller(diff)[1])
```
